```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
h = 6.63e-34  # Planck's constant (Joule seconds)
c = 3.00e8  # Speed of light (meters/second)
k_B = 1.38e-23  # Boltzmann constant (Joules/Kelvin)
T = 5800  # Temperature of the Sun (Kelvin)

# Wavelength range (in meters)
wavelengths = np.linspace(1e-9, 3e-6, 1000)

# Planck's law function
def planck(wavelength, T):
    return (2 * np.pi * h * c**2) / (wavelength**5 * (np.exp((h * c) / (wavelength * k_B * T)) - 1))

# Calculate power per unit wavelength
power = planck(wavelengths, T)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(wavelengths * 1e9, power, label=f'T = {T} K')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Spectral Radiance (W/m^3)')
plt.title('Blackbody Radiation Spectrum')
plt.legend()
plt.grid(True)
plt.xlim(0, 3000)
plt.ylim(0, max(power)*1.1)
plt.axvline(x=500, color='r', linestyle='--')
plt.text(550, max(power)*0.9, 'lambda_max ~ 500 nm', color='r')
plt.savefig("fig3.png", dpi=500)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
h = 6.63e-34  # Planck's constant (Joule seconds)
c = 3.00e8  # Speed of light (meters/second)
k_B = 1.38e-23  # Boltzmann constant (Joules/Kelvin)
T = 255  # Temperature of the Earth (Kelvin)

# Wavelength range (in meters)
wavelengths = np.linspace(1e-6, 100e-6, 1000)  # Adjust range for infrared

# Planck's law function
def planck(wavelength, T):
    return (2 * np.pi * h * c**2) / (wavelength**5 * (np.exp((h * c) / (wavelength * k_B * T)) - 1))

# Calculate power per unit wavelength
power = planck(wavelengths, T)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(wavelengths * 1e6, power, label=f'T = {T} K')
plt.xlabel('Wavelength (micrometers)')
plt.ylabel('Spectral Radiance (W/m^3)')
plt.title('Blackbody Radiation Spectrum')
plt.legend()
plt.grid(True)
plt.xlim(0, 100)
plt.ylim(0, max(power)*1.1)
plt.axvline(x=11.4, color='r', linestyle='--')
plt.text(15, max(power)*0.9, 'lambda_max ~ 11.4 micrometers', color='r')
plt.savefig("fig4.png", dpi=500)
plt.show()
```
